### Starbucks vs. Chipotle vs. Shake Shack

Stock analysis for Starbucks, Chipotle, and Shake Shack from August 6, 2018 to August 6, 2019. Stock data downloaded from Yahoo! Finance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader
from datetime import datetime
import pandas_datareader.data as web
```

```python
#Starbucks
starbucks = pd.read_csv('SBUX.csv', parse_dates=True)
starbucks['Date'] = pd.to_datetime(starbucks['Date'])
starbucks = starbucks.set_index('Date')
starbucks
```

```python
#Chipotle
chipotle = pd.read_csv('CMG.csv', parse_dates=True)
chipotle['Date'] = pd.to_datetime(chipotle['Date'])
chipotle = chipotle.set_index('Date')
```

```python
chipotle
```

```python
#Shake Shack
sshack = pd.read_csv('SHAK.csv', parse_dates=True)
sshack['Date'] = pd.to_datetime(sshack['Date'])
sshack = sshack.set_index('Date')
sshack
```

### Opening Prices

```python
starbucks['Open'].plot(label='Starbucks', figsize=(10,9),title='Opening Prices')
chipotle['Open'].plot(label='Chiptole')
sshack['Open'].plot(label='Shake Shack')
plt.legend();
```

### Volume of Stock Traded Each Day

```python
starbucks['Volume'].plot(label='Starbucks', figsize=(20,9),title='Volume Traded Each Day')
chipotle['Volume'].plot(label='Chiptole')
sshack['Volume'].plot(label='Shake Shack')
plt.legend();
```

Starbucks had a major spike in volume traded on November 2, 2018. November 2, 2018 was also the day that Starbucks gave our free, reusable red cups to anyone who purchased a holiday drink, as long as the red cups were in stock in the store. Between November 3, 2018, and January 7, 2019, anyone who brought in the red cup after 2pm received 50 cents off their grande drink order.

```python
starbucks['Volume'].idxmax()
```

```python
starbucks['Total Traded'] = starbucks['Open']*starbucks['Volume']
chipotle['Total Traded']  = chipotle['Open']*chipotle['Volume']
sshack['Total Traded'] = sshack['Open']*sshack['Volume']
```

```python
starbucks['Total Traded'].plot(label='Starbucks', figsize=(20,9), title='Total Traded')
chipotle['Total Traded'].plot(label='Chipotle', figsize=(20,9))
sshack['Total Traded'].plot(label='Shake Shack', figsize=(20,9))
plt.legend()
```

On February 7, 2019, the amount of total traded for Chipotle spiked because on February 6, 2019, the company released their fourth quarter results, which improved greatly almost across the entire board.

### Daily Percent Change¶
Percent gain or percent loss.

More volatile (not centered around 0%) = higher risk, higher reward

```python
starbucks['Returns'] = starbucks['Close'].pct_change(1) #1 is day to day
chipotle['Returns'] = chipotle['Close'].pct_change(1) #1 is day to day
sshack['Returns'] = sshack['Close'].pct_change(1) #1 is day to day
```

```python
starbucks['Returns'].hist(bins=100, label='Starbucks', figsize=(10,8))
```

```python
chipotle['Returns'].hist(bins=100, label='Chipotle', figsize=(10,8))
```

```python
sshack['Returns'].hist(bins=100, label='Shake Shack', figsize=(10,8))
```

```python
starbucks['Returns'].hist(bins=100, label='Starbucks', figsize=(10,8),alpha=0.5)
chipotle['Returns'].hist(bins=100, label='Chipotle', figsize=(10,8), alpha=0.5)
sshack['Returns'].hist(bins=100, label='Shake Shack', figsize=(10,8), alpha=0.5)
plt.legend()
```

```python
starbucks['Returns'].plot(kind='kde', label='Starbucks', figsize=(10,8))
chipotle['Returns'].plot(kind='kde', label='Chipotle', figsize=(10,8))
sshack['Returns'].plot(kind='kde', label='Shake Shack', figsize=(10,8))
plt.legend()
```

```python
box_df = pd.concat([starbucks['Returns'], chipotle['Returns'], sshack['Returns']], axis=1)
box_df.columns = ['Starbucks Returns', 'Chipotle Returns', 'Shake Shack Returns']
```

```python
box_df.plot(kind='box', figsize=(10,14))
```

### Cumulative Daily Returns

```python
starbucks['Cumulative Return'] = (1+starbucks['Returns']).cumprod()
starbucks
```

```python
chipotle['Cumulative Return'] = (1+chipotle['Returns']).cumprod()
chipotle
```

```python
sshack['Cumulative Return'] = (1+sshack['Returns']).cumprod()
sshack
```

```python
starbucks['Cumulative Return'].plot(label='Starbucks', figsize=(10,10))
chipotle['Cumulative Return'].plot(label='Chipotle', figsize=(10,10))
sshack['Cumulative Return'].plot(label='Shake Shack', figsize=(10,10))
plt.legend()
```

Starbucks has the largest cumulative return.

Shake Shack had the lowest cumulative return.
